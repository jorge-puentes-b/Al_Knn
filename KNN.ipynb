```python
# Se importan librerias, el modelo KNN y las visualizaciones
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, accuracy_score
from sklearn.datasets import load_iris

# Paso 2: Cargar el conjunto de datos (Iris en este ejemplo)
# Utilizamos el conjunto de datos Iris proporcionado por scikit-learn
iris = load_iris()
X = iris.data  # Características (atributos)
y = iris.target  # Etiquetas (clases)

# Paso 3: Dividir el conjunto de datos en entrenamiento y prueba
# Dividimos el dataset en un 70% para entrenamiento y 30% para prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Paso 4: Entrenar el modelo usando el algoritmo KNN
# Inicializamos el clasificador KNN con k=3 (puedes ajustar este valor)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Paso 5: Evaluar el rendimiento del modelo
# Realizamos predicciones sobre el conjunto de prueba
y_pred = knn.predict(X_test)

# Calculamos y mostramos el reporte de clasificación (precision, recall, F1-score)
print("Reporte de Clasificación:\n", classification_report(y_test, y_pred))

# Calcular y mostrar la precisión
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy:.2f}")

# Paso 6: Matriz de Confusión
# La matriz de confusión nos permite ver la cantidad de aciertos y errores para cada clase
conf_matrix = confusion_matrix(y_test, y_pred)

# Graficar la matriz de confusión
plt.figure(figsize=(6,6))
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt='g', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Matriz de Confusión")
plt.xlabel("Predicciones")
plt.ylabel("Valores Reales")
plt.show()

# Paso 7: Curva ROC (solo si es problema binario, en este caso no aplica ya que Iris tiene 3 clases)
# Si estuvieras usando un dataset binario, podrías graficar la curva ROC.
# Ejemplo para problemas binarios:
# fpr, tpr, thresholds = roc_curve(y_test_binary, y_prob)
# plt.plot(fpr, tpr)
# plt.title('Curva ROC')
# plt.xlabel('False Positive Rate')
# plt.ylabel('True Positive Rate')
# plt.show()

# Nota: Puedes usar un dataset binario como el de dígitos MNIST para aplicar ROC en problemas de dos clases.
```
